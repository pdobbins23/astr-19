```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 100
z = np.arange(n, dtype=float)
z *= 2.0 * np.pi / float(n - 1)
sin_z = np.sin(z)
```

```python
# Make a plot
plt.plot(z, sin_z, color="red")
plt.xlabel("z")
plt.ylabel("sin(z)")
```

```python
x = np.arange(0, 5, 0.1)
y = np.sin(x)

plt.plot(x, y, color="red")
plt.xlabel("x")
plt.ylabel("sin(x)")

plt.savefig("sinx.png", bbox_inches="tight", dpi=600)
```

```python
x = np.arange(10) + 2*np.random.randn(10)
y = np.arange(10) + 2*np.random.randn(10)

xerr = 2*np.random.random(10)
yerr = 2*np.random.random(10)

plt.errorbar(x, y, xerr=xerr, yerr=yerr, fmt="bo")
```

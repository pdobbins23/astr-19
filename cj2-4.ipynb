```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.gamma(1, 2, 1000)

width = 0.01

bins = np.arange(0, 1, width)
plt.hist(x, bins=bins, alpha=0.5, edgecolor="black")
plt.ylabel("N per bin")
plt.xlabel("x")
plt.savefig("cj2-4.pdf", bbox_inches="tight", facecolor="white")
```
